### Displaying Currencies

```
df_currency_all = df['currency_name'].value_counts()
allowed_currencies = df_currency_all.head(3).index
df_currency_all = df_currency_all.groupby(df_currency_all.index.map(lambda x: x if x in allowed_currencies else "Others")).sum()

df_egypt = df[df['country'] == 'Egypt']
df_currency_egypt = df_egypt['currency_name'].value_counts()
df_currency_egypt = df_currency_egypt.groupby(df_currency_egypt.index.map(lambda x: x if x in allowed_currencies else "Others")).sum()

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Whole Site Chart
wedges1, texts1, autotexts1 = axes[0].pie(
    df_currency_all, labels=None, autopct='%1.1f%%',
    startangle=180, wedgeprops={'edgecolor': 'black'},
    textprops={'fontsize': 12}, pctdistance=0.8, labeldistance=1.1
)
axes[0].set_title('Currency Distribution in Job Postings (Whole Site)', fontsize=14, fontweight='bold')
axes[0].legend(wedges1, df_currency_all.index, title="Currency", loc="upper right")

# Egypt-Specific Chart
wedges2, texts2, autotexts2 = axes[1].pie(
    df_currency_egypt, labels=None, autopct='%1.1f%%',
    startangle=180, wedgeprops={'edgecolor': 'black'},
    textprops={'fontsize': 12}, pctdistance=0.8, labeldistance=1.1
)
axes[1].set_title('Currency Distribution in Job Postings (Egypt)', fontsize=14, fontweight='bold')
axes[1].legend(wedges2, df_currency_egypt.index, title="Currency", loc="upper right")

plt.tight_layout()
plt.show()
```
